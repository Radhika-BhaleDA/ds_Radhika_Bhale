```python
from google.colab import files
import pandas as pd
import io

# Upload CSV file
uploaded = files.upload()

# Read the uploaded file
filename = list(uploaded.keys())[0]  # get file name automatically
df = pd.read_csv(io.BytesIO(uploaded[filename]))

df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Timestamp IST'] = pd.to_datetime(df['Timestamp IST'], format='%d-%m-%Y %H:%M')
```

```python
df.info()
```

```python
df.drop(columns=['Timestamp'], inplace=True)
```

```python
df.info()
```

```python
# Copy the dataframe (optional but safe)
df_clean = df.copy()

# 1. Extract date from timestamp
df_clean['Date'] = df_clean['Timestamp IST'].dt.date

# 2. Extract hour (helpful for intraday patterns)
df_clean['Hour'] = df_clean['Timestamp IST'].dt.hour

# 3. Create Profit Status (Win/Loss)
df_clean['PnL_Result'] = df_clean['Closed PnL'].apply(lambda x: 'Profit' if x > 0 else 'Loss')

# 4. Create absolute profit for comparison
df_clean['Closed PnL Abs'] = df_clean['Closed PnL'].abs()

df_clean.head()
```

```python
from google.colab import files
import pandas as pd
import io

uploaded2 = files.upload()

filename2 = list(uploaded2.keys())[0]
sentiment = pd.read_csv(io.BytesIO(uploaded2[filename2]))

sentiment.head()
```

```python
sentiment['date'] = pd.to_datetime(sentiment['date'], errors='coerce')
sentiment['date'] = sentiment['date'].dt.date
```

```python
sentiment.info()
```

```python
merged_df = df_clean.merge(sentiment, left_on="Date", right_on="date", how="left")
merged_df.head()
```

```python
import matplotlib.pyplot as plt
import os

# Create the directory if it doesn't exist (run once at the top)
os.makedirs('outputs', exist_ok=True)

plt.figure(figsize=(6,4))
merged_df['classification'].value_counts().plot(kind='bar')
plt.title('Market Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Trading Records')

# Save the chart BEFORE show
plt.savefig("outputs/market_sentiment_distribution.png")

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(data=merged_df, x='classification', hue='Side')
plt.title('Buy vs Sell Activity Under Different Sentiments')
plt.xlabel('Market Sentiment')
plt.ylabel('Trade Count')
plt.xticks(rotation=20)


# Save the chart
plt.savefig("outputs/Buy vs Sell Activity Under Different Sentiments.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=merged_df, x='classification', hue='PnL_Result')
plt.title('Profit vs Loss Distribution Across Sentiment States')
plt.xlabel('Market Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=20)


# Save the chart
plt.savefig("outputs/Profit vs Loss Distribution Across Sentiment States")

plt.show()
```

```python
volume = merged_df.groupby('classification')['Size USD'].sum()

plt.figure(figsize=(6,4))
volume.plot(kind='bar')
plt.title('Total Trading Volume by Market Sentiment')
plt.ylabel('Total Size USD')
plt.xlabel('Sentiment')


# Save the chart
plt.savefig("outputs/Total Trading Volume by Market Sentiment.png")

plt.show()
```
